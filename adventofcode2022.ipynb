```python
import numpy as np
with open("input1.txt") as f:
    lines = f.read()

elfs = lines.split("\n\n")

i=0
s = []
for elf in elfs:
  calories = elf.split("\n")

  if('' in calories):
    calories.remove('')
  res = np.asarray(calories,dtype = int)

  suma = np.sum(res)

  s.append(suma)
  if suma > i:
    i = suma

sortedList = sorted(s)
print("Part 1:",  sortedList[-1])
print("Part 2:", np.sum( sortedList[-3:]))
```

```python
def score(p1, p2):

  if (p1 - 1) % 3 == p2:
    return p2 + 1
  elif (p2 - 1) % 3 == p1:
    return p2 + 7
  return p2 + 4

with open("input2.txt", "r") as f:

  data = f.read().splitlines()
  data = [line.split() for line in data]
  data = [(ord(p1) - ord("A"), ord(p2) - ord("X")) for p1, p2 in data]

moves = [(p1, (p1 + p2 - 1) % 3) for p1, p2 in data]

print("Part 1:", sum(score(p1, p2) for p1, p2 in data))
print("Part 2:", sum(score(p1, p2) for p1, p2 in moves))
```

```python
print("Advent of Code 2022 - Day 3")
import string

f = open("./input3.txt", "r");
lines = f.readlines();

bags = []
suma = 0;

for line in lines:
    bags.append(line.replace("\n", ""))

for bag in bags:
    #print(bag)
    compartment1, compartment2 = bag[:len(bag)//2], bag[len(bag)//2:]
    #print(compartment1)
    #print(compartment2)

    a=list(set(compartment1)&set(compartment2))
    #print("The common letters are:")
    for i in a:
        #print(i)
        suma += string.ascii_letters.index(i)+1

print("Priority:", suma)

# 8088
```
